```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/laptops.csv',encoding='ISO-8859-1')
df
```

```python
df.info()
```

```python
#To get dimensions of the dataset
df.shape
```

```python
#To find null values
df.isnull().sum()
```

```python
#to drop/remove the Unnamed: 0 column
df=df.drop(columns = 'Unnamed: 0')
df
```

```python
#Removeing the 'GB' from the Ram column and keeping only the integer value
df.Ram=df.Ram.str.replace('GB',' ')
df
```

```python
#converting the data of the Ram columns from string to int
df.Ram = df.Ram.astype('int64')
df.info()
```

```python
#removing the kg from Weight column and keeping only the integer
df.Weight=df.Weight.str.replace('kg',' ')
df.Weight=df.Weight.astype('float64')
df.info()
```

```python
#VISUALIZATION
import seaborn as sns
sns.distplot(df.Price_euros)
```

```python
#To find out which brand laptops are more in number
df.Company.value_counts()
```

```python
#To make graph of above obtained data'
df.Company.value_counts().plot(kind='bar')
```
